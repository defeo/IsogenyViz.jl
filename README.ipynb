# IsogenyViz: visualizing isogeny graphs

This package produces visualizations of isogeny graphs of elliptic curves. It uses [Luxor](https://juliagraphics.github.io/Luxor.jl/) for graphics, and [Nemo](nemocas.org/) and [EllipticCurves.jl](github.com/defeo/EllipticCurves.jl/) for constructing the isogeny graphs.

Right now, the following isogeny graphs are supported:

- *Complex Multiplication graphs*, or Cayley graphs of class groups, via the `CMgraph` function. These do not use any actual data, they simply draw an abstract Cayley graph.

All functions produce "unwrapped" Luxor plotting material. It is up to you to generate the full picture using any of the Luxor macros `@draw`, `@png`, `@svg`, or more generic Luxor commands. See the [Luxor manual](https://juliagraphics.github.io/Luxor.jl/).

See the examples below.

```julia
using Pkg
Pkg.activate(".")
```

```julia
using Luxor
using IsogenyViz
```

## Complex Multiplication graphs

The function `CMgraph` takes two parameters:

- `nodes::Int`, a number of vertices, and
- `arcs::Vector{Int}`, a list of "jumps" defining the edges.

Other parameters controlling the appeareance are: `sagittas`, `colors` and `scale`. See below.

```julia
@doc CMgraph
```

### A simple example: 13 vertices, edges jump by 1, 3 and 5 vertices.

The output is stored in memory as a PNG, and shown interactively (right click + save image to download)

```julia
@draw CMgraph(13, [1, 3, 5])
```

### A full (cute) example

We use `sagittas` to control the bending of the arcs, we choose a different color palette, and we make a larger picture (600px diameter, in a box of 650px×650px)

```julia
@draw CMgraph(31, [1, 5, 11, 17];
    sagittas=[0.01, -0.1, 0.02, 0.02],
    colors=["magenta", "orange", "cyan", "silver"],
    scale=600) 650 650
```

### A large example

This is saved to a file named `large.svg`, which you can download from the [file explorer](/tree)

```julia
@svg CMgraph(101, [1, 11, 21, 57];
    sagittas=[0.005,-0.01,-0.1,0.05],
    colors=["blue", "gold", "magenta", "red"],
    scale=900) 1000 1000 "large.svg"
```
